```python
for dataset in data_cleaner:
    dataset['Age'].fillna(dataset['Age'].median(), inplace= True)
    dataset.dropna(subset= ['Embarked', 'Fare'], inplace= True)

drop_columns= ['PassengerId', 'Cabin', 'Ticket']
data1.drop(drop_columns, axis= 1, inplace= True)

print(data1.isnull().sum())
print("-"*10)
print(data_val.isnull().sum())
```

```python
for dataset in data_cleaner:
    dataset['FamilySize']= dataset['SibSp']+ dataset['Parch']+ 1
    dataset['IsAlone']= 1
    dataset['IsAlone'].loc[dataset['FamilySize'] >1]= 0

    dataset['Title'] = dataset['Name'].str.split(', ').str[1].str.split('.').str[0]

    dataset['FareBin']= pd.qcut(dataset['Fare'], 4)
    dataset['AgeBin']= pd.cut(dataset['Age'].astype(int), 5)

min_nb= 10
is_rare= (data1['Title'].value_counts() < min_nb)

data1['Title']= data1['Title'].apply(lambda x: 'Misc' if is_rare.loc[x] == True else x)
print(data1['Title'].value_counts())

print("-"*10)

data1.head()
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# Grille des hyperparamtres  tester
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [4, 6, 8, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Dfinition de la validation croise
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Configuration de la recherche en grille
grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(),  # Instanciation du modle
    param_grid=param_grid,
    cv=cv,
    scoring='accuracy',
    n_jobs=-1  # Utilise tous les curs pour acclrer
)

# Entranement de la recherche en grille
grid_search.fit(X_train, y_train)

# Rsultats de la recherche
print("Meilleurs paramtres :", grid_search.best_params_)
print("Meilleure accuracy sur le training set :", grid_search.best_score_)

model = grid_search.best_estimator_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# 1. Dfinir les modles et leurs grilles de paramtres
models_and_params = {
    'DecisionTree': {
        'model': DecisionTreeClassifier(random_state=42),
        'params': {
            'criterion': ['gini', 'entropy'],
            'max_depth': [4, 6, 8, None],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    'RandomForest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200],
            'max_depth': [4, 6, 8, None],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    'SVM': {
        'model': SVC(random_state=42),
        'params': {
            'C': [0.1, 1, 10],
            'kernel': ['linear', 'rbf', 'poly'],
            'gamma': ['scale', 'auto']
        }
    }
}

# 2. Initialisation de la validation croise
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 3. Stockage des rsultats
best_models = {}
results = []

for name, config in models_and_params.items():
    print(f"Optimisation pour : {name}")
    grid_search = GridSearchCV(
        estimator=config['model'],
        param_grid=config['params'],
        cv=cv,
        scoring='accuracy',
        n_jobs=-1,
        verbose=1
    )
    
    # Entranement du modle
    grid_search.fit(X_train, y_train)
    
    # Sauvegarder les rsultats
    best_models[name] = grid_search.best_estimator_
    results.append({
        'Model': name,
        'Best Params': grid_search.best_params_,
        'Best Score': grid_search.best_score_
    })

# 4. Afficher les rsultats
results_df = pd.DataFrame(results)
print("\nRsum des rsultats :")
print(results_df)

# 5. valuer le meilleur modle sur X_test
best_model_name = results_df.loc[results_df['Best Score'].idxmax(), 'Model']
best_model = best_models[best_model_name]

print(f"\nMeilleur modle : {best_model_name}")
print(classification_report(y_train, best_model.predict(X_train)))
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score

# 1. Prparation des donnes
# Identifier les colonnes numriques et catgoriques
numeric_features = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X_train.select_dtypes(include=['object']).columns

# Prtraitement pour les colonnes numriques et catgoriques
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent'))
])

# Appliquer les transformations
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# 2. Pipeline avec le modle
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(random_state=42))
])

# 3. Optimisation des hyperparamtres
param_grid = {
    'classifier__criterion': ['gini', 'entropy'],
    'classifier__max_depth': [4, 6, 8, 10, None],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring='accuracy', n_jobs=-1)

# Entranement
grid_search.fit(X_train, y_train)

# 4. Meilleurs paramtres et valuation
print("Meilleurs paramtres :", grid_search.best_params_)
print("Meilleure accuracy sur le training set :", grid_search.best_score_)

# 5. Prdictions sur le test set
y_pred = grid_search.best_estimator_.predict(X_test)

# Exemple d'valuation (si les vraies valeurs y_test sont disponibles)
# print("Accuracy sur le test set :", accuracy_score(y_test, y_pred))
```

```python
# 1. Dfinir les modles et leurs grilles de paramtres
models_and_params = {
    'DecisionTree': {
        'model': DecisionTreeClassifier(random_state=42),
        'params': {
            'criterion': ['gini', 'entropy'],
            'max_depth': [4, 6, 8, None],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    'RandomForest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200],
            'max_depth': [4, 6, 8, None],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    'SVM': {
        'model': SVC(random_state=42),
        'params': {
            'C': [0.1, 1, 10],
            'kernel': ['linear', 'rbf', 'poly'],
            'gamma': ['scale', 'auto']
        }
    }
}

# 2. Initialisation de la validation croise
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 3. Stockage des rsultats
best_models = {}
results = []

for name, config in models_and_params.items():
    print(f"Optimisation pour : {name}")
    grid_search = GridSearchCV(
        estimator=config['model'],
        param_grid=config['params'],
        cv=cv,
        scoring='accuracy',
        n_jobs=-1,
        verbose=1
    )
    
    # Entranement du modle
    grid_search.fit(X_train, y_train)
    
    # Sauvegarder les rsultats
    best_models[name] = grid_search.best_estimator_
    results.append({
        'Model': name,
        'Best Params': grid_search.best_params_,
        'Best Score': grid_search.best_score_
    })

# 4. Afficher les rsultats
results_df = pd.DataFrame(results)
print("\nRsum des rsultats :")
print(results_df)

# 5. valuer le meilleur modle sur X_test
best_model_name = results_df.loc[results_df['Best Score'].idxmax(), 'Model']
best_model = best_models[best_model_name]

print(f"\nMeilleur modle : {best_model_name}")
print(classification_report(y_train, best_model.predict(X_train)))
```
